```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel(r"C:\Users\Preeti\Desktop\Data\Salary.xlsx")
df
```

```python
inputs=df.drop("Salary_more_than_100k",axis='columns')
inputs
```

```python
targets=df["Salary_more_than_100k"]
targets
```

```python
from sklearn.preprocessing import LabelEncoder
le_company= LabelEncoder()
le_Job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
inputs["Company_n"]=le_company.fit_transform(inputs["Company"])
inputs["degree"]=le_degree.fit_transform(inputs["Degree"])
inputs["job"]=le_Job.fit_transform(inputs["Job "])
```

```python
inputs
```

```python
new_input=inputs.drop(["Company","Job ","Degree"],axis="columns")
```

```python
new_input
```

```python
targets
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(new_input,targets)
```

```python
model.score(new_input,targets)
```

# Is salary of Google, Computer Engineer, Bachelors degree > 100 k ?
